# Tipos de aprendizaje
- Datos estructurados: Aprendizaje automático clásico
- Datos complejos, imágenes, etc: Aprendizaje profundo con redes neuronales
# Tipos de problemas
- Clasificación: Tenemos datos que queremos clasificar en varias clases. Nos basamos en datos de clasificaciones previas (conjunto de entrenamiento).
- Regresión: En lugar de clasificar en clases discretas, asignamos un valor continuo como salida al dato de entrada.
- Aprendizaje no supervisado: No tenemos una clasificación previa para realizar el entrenamiento. Se agrupan datos de tipos similares (segmentación/clustering). Reducción de dimensionalidad (ver datos que no influyen o son redundantes).


```python
import sklearn as sk
from sklearn import datasets
import pandas as pd
```

```python
iris = sk.datasets.load_iris()
```

```python
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
df['target'] = iris['target']
df.value_counts()
```

```python
import seaborn as sb
sb.pairplot(df, hue = 'target')
```

```python
from sklearn import neighbors
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

# Ejemplo sin separación de los datos de entrenamiento y de validación

```python
knh = neighbors.KNeighborsClassifier(n_neighbors=5)
x = iris['data']
y = iris['target']
knh.fit(x, y)
```

```python
y_pred = knh.predict(x)
```

```python
acc = accuracy_score(y, y_pred)
print(f'Porcentaje de acierto de {acc*100:.2f}%')
```

# Ejemplo con separación de datos

```python
knh_2 = neighbors.KNeighborsClassifier(n_neighbors=5)
```

```python
x = iris['data']
y = iris['target']
X_train, X_test, Y_train, Y_test = train_test_split(x,y)
```
